```python
data = pd.read_csv("testData3.csv",index_col = False)
```

```python
data
```

```python
data['Hướng ban công'].isnull().sum()
```

```python
data = data[data['Hướng ban công'].isnull() == False]
```

```python
attributes = list(['Diện tích','Số phòng ngủ','Số toilet'])#,'Hướng nhà','Hướng ban công','Chủ đầu tư','Vĩ độ','Kinh độ'])
X = data[attributes]
```

```python
y = data['Giá']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(np.array(X), np.array(y), test_size=0.3,random_state = 42)
```

```python
from sklearn import linear_model
linear = linear_model.LinearRegression()
```

```python
linear.fit(X_train,y_train)
```

```python
linear.score(X_train,y_train)
```

```python
linear.score(X_test,y_test)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
```

```python
poly = PolynomialFeatures(degree= 3)
X_trans = poly.fit_transform(X_train)
```

```python
linear.fit(X_trans,y_train)
```

```python
X_trans_test = poly.fit_transform(X_test)
y_trans_pred = linear.predict(X_trans_test)
```

```python
linear.score(X_trans,y_train)
```

```python
linear.score(X_trans_test,y_test)
```

```python
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_trans_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_trans_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_trans_pred)))
```

```python
import pickle

pickle.dump(linear, open('poly_regression_2_10.pkl', 'wb'))
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestRegressor

regressor = RandomForestRegressor(n_estimators=50, random_state=0)
regressor.fit(X_train, y_train)
y_pred = regressor.predict(X_test)
```

```python
regressor.score(X_test,y_test)
```

```python
from sklearn import metrics

print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
