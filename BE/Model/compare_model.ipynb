```python
attributes = list(['DienTich','SoPhongNgu','SoToilet','HuongNha','HuongBanCong','ChuDauTu','ViDo','KinhDo','Quan','LoaiTin'])
X = data[attributes]
```

```python
y = data['GiaTien']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import Pipeline
```

```python
poly = PolynomialFeatures(degree= 3)
X_trans = poly.fit_transform(X_train)
```

```python
X_trans_test = poly.fit_transform(X_test)
```

```python
import pickle
model = pickle.load(open('random_forest_20_11.pkl','rb'))
```

```python
y_pred = model.predict(X_test)
```

```python
model.score(X_test,y_test)
```

```python
model.score(X_train,y_train)
```

```python
len(y_pred)
```

```python
plt.figure(figsize=(25,7))
plt.title("Random Forest")
sns.distplot(y_pred,kde=True,hist = False,color='blue')
sns.distplot(y_test,kde=True,hist = False,color='green')
plt.show()
```

```python
plt.figure(figsize=(25,7))
plt.title("KNN")
sns_plot = sns.distplot(y_pred,kde=True,hist = False,color='blue')
sns_plot = sns.distplot(y_test,kde=True,hist = False,color='green')
plt.show()
sns_plot.savefig("output.png")
```
