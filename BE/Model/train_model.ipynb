```python
data = data[data['Price']<5.5e10]
```

```python
data.drop(data[(data['Area'] > 250) & (data['Price']<1.5e10)].index,inplace=True)
```

```python
attributes = list(['Area','NumberBed','NumberToilet','Chudautu','ViDo','KinhDo','NoiThat','LoaiTin','Quan','HuongNha','HuongBanCong'])
X = data[attributes]
```

```python
y = data['Price']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(np.array(X), np.array(y), test_size=0.3,random_state = 42)
```

```python
from sklearn import linear_model
linear = linear_model.LinearRegression()
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
linear.fit(X_train,y_train)
```

```python
linear.score(X_train,y_train)
```

```python
linear.score(X_test,y_test)
```

```python
y_pred = linear.predict(X_test)
```

```python
df = pd.DataFrame({'Actual': y_test, 'Predicted': y_pred})
```

```python
plt.figure(figsize = (15,7))
plt.plot(df)
plt.show()
```

```python
plt.figure(figsize=(15, 7))
sns.distplot(y_pred, kde=True, hist = False, color= 'blue')
sns.distplot(y_test,kde=True, hist = False, color= 'green')
plt.show()
```
