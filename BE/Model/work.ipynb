```python
import pandas as pd 
import numpy as np
import re
```

```python
data = pd.read_csv("testData4.csv")
```

```python
data
```

```python
train_df = pd.DataFrame(data)
```

```python
train_df.drop(columns = ['title','MoTa','NgayDang','Phone','url','Image'],axis = 1,inplace = True)
```

```python
train_df
```

```python
train_df['DienTich'] = train_df['DienTich'].str.extract('(\d*\.\d+|\d+)').astype(float)
```

```python
extract_price = train_df[train_df['GiaTien'].str.contains('m²') == True]['GiaTien'].str.extract('(\d*\.\d+|\d+)').astype(float)[0] * train_df[train_df['GiaTien'].str.contains('m²') == True]['DienTich']
```

```python
extract_price = pd.DataFrame({'index':extract_price.index,'price':extract_price.values})
```

```python
extract_price['price'] = extract_price['price']*1e6
```

```python
extract_price
```

```python
lst = train_df[train_df['GiaTien'].str.contains('m²') == True]['GiaTien'] 
lst = lst.to_list()
lst_1 = extract_price['price']
lst_1 = lst_1.to_list()
```

```python
train_df['GiaTien'] = train_df['GiaTien'].replace(lst,lst_1)
```

```python
train_df = train_df[train_df['GiaTien'].str.contains('Thỏa thuận') == False]
```

```python
train_df
```

```python
train_df['GiaTien'] = train_df['GiaTien'].str.replace(' tỷ','e+9').str.replace(' triệu','e+6').astype('float')
```

```python
train_df
```

```python
train_df['SoPhongNgu'] = train_df['SoPhongNgu'].str.extract('(\d+)').astype(int)
```

```python
train_df
```

```python
train_df['SoToilet'].unique()
```

```python
train_df['SoToilet'] = train_df['SoToilet'].str.replace('((phòng))','')
```

```python
train_df['SoToilet'] = train_df['SoToilet'].str.replace(')','',regex=True)
```

```python
train_df['SoToilet'] = train_df['SoToilet'].str.replace('(','',regex=True)
```

```python
train_df['SoToilet'].fillna(0,inplace= True)
```

```python
train_df['SoToilet'] = train_df['SoToilet'].astype(int)
```

```python
train_df
```

```python
train_df.to_csv("data_clean_for_train_NDT.csv",index = False)
```
